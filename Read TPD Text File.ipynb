```python
import pandas as pd

fileName = 'TPD220722_01.txt'
fileLocation = 'C:/Users/larue/Chapman University/LaRue CatLab - Data/2022/2022.07.22/'

data = pd.read_csv(fileLocation+fileName,delimiter='\t')
data

newheader = list()
for name in data.columns :
    newname = name.replace('TPD220722_01_','')
    if newname == 'ch1' :
        newname = 'Temp'
    newheader.append(newname)
data.columns=newheader

import matplotlib.pyplot as plt
import seaborn as sns

# Format plot
fontsize = 22
sns.set_theme(style="ticks")
fig, ax = plt.subplots(figsize=(20,8))

ax.set_xlabel('Temperature (K)',fontsize=fontsize)
ax.set_ylabel('Intensity (au)',fontsize=fontsize)
ax.tick_params(axis = 'both', which = 'both', labelsize = fontsize)

for idx, col in enumerate(data) :
    if idx > 1 :
        plt.plot(data['Temp'][2:],data[col][2:],label=col)
plt.legend(frameon=False,fontsize=fontsize-6)
plt.show()
```
